```python
class Crawler:
    
    def __init__(self):
        self.to_visit = list()
        self.visited = set()
        self.data = []

        
    def fetch(self, url):
        
        print('now fetching.. ' , url)
        res = requests.get(url)
        return res.content  
    
    def get_current_url(self):
        res = self.to_visit.pop()
        while res is self.visited:
            res = self.to_visit.pop()
        return res
    
                
    
    def get_links(self, content):
        urls = re.findall('<a href="([^"]+)">',str(content))
        for url in urls:
            pattern = re.compile('http?')
            if pattern.match(url):
                self.to_visit.append(url)
    
    def crawl(self, url, depth=35):
        self.to_visit.append(url)
        while len(self.visited) <= depth:
            current_url = self.get_current_url()
            if current_url not in self.visited:
                if 'privacy-policy.html' not in current_url:
                    content = self.fetch(current_url)
                
                    self.visited.add(current_url)
                    self.get_links(content)
                
                    # create data
                    s = BeautifulSoup(content, 'html.parser')  
                    d = {}
                    br = ". "
                    
                    # fetch title
                    title = ""
                    for a in s.find_all('title'):
                        title += a.get_text().strip() + br
                
                    # fetch h2
                    h2 = ""
                    for a in s.find_all('h2'):
                        h2 += a.get_text().strip() + br
                    
                    # fetch td
                    td = ""
                    for a in s.find_all('td'):
                        td += a.get_text().strip() + br
                
                    # fetch paragraph
                    paragraphs = ""
                    for a in s.find_all('p'):
                        paragraphs += a.get_text().strip() + br
                    
                    d['url'] = current_url
                    d['title'] = title
                    d['h2'] = h2
                    d['td'] = td
                    d['paragraph'] = paragraphs
                    self.data.append(d)
```
